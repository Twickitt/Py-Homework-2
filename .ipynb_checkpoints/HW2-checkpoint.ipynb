## Задача Numpy 1 (10 баллов)

Создать массивы и вывести на экран:
- содержимое
- форму
- размер занимаемой памяти

1. одномерный массив, содержащий действительные числа $\pi, \frac{2}{\pi}, \frac{\pi}{4}, \frac{8}{\pi}, \frac{\pi}{16}, ..., \frac{\pi}{1024}$
2. одномерный массив, содержащий константу `True`, если число делится на 3, но не делится на 6, иначе - `False` для чисел 1, ..., 101.
3. двумерный массив, содержащий аргументы комплексных чисел $x + iy$, где $x = 2k/3, y=5m/3, k=0,...,10, m=0,...,10$

Оценивание:
- [5 баллов] использовать генераторы списков
- [5 баллов] использовать только возможности `Numpy`

````python
import numpy as np


np.set_printoptions(suppress=True)


P = np.pi
arr1 = np.array([P, 2/P, P/4, 8/P, P/16, 32/P, P/64, 128/P, P/256, 512/P, P/1024], dtype=float)

print("1:\nМассив:\n", np.round(arr1, 4))
print("Форма:", arr1.shape)
print("Размер (в байтах):", arr1.nbytes)


arr2 = np.array([(x % 3 == 0 and x % 6 != 0) for x in range(1, 102)], dtype=bool)

print("\n2:\nМассив:\n", arr2)
print("Форма:", arr2.shape)
print("Размер (в байтах):", arr2.nbytes)


k_vals = [2 * k / 3 for k in range(11)]
m_vals = [5 * m / 3 for m in range(11)]

complex_grid = np.array([[complex(x, y) for y in m_vals] for x in k_vals])
arg_array = np.angle(complex_grid)

print("\n3:\nМассив аргументов:\n", np.round(arg_array, 4))
print("Форма:", arg_array.shape)
print("Размер (в байтах):", arg_array.nbytes)
````

- $y = cos(t)(e^{cos(t)} - 2cos(4t) + sin^5(\frac{t}{12}))$

## Задача Matplotlib 3 (10 баллов)

**Требуется**

1. самостоятельно придумать непрерывную функцию $f(x, y)$, нелинейно зависящую от $x$ и $y$.
2. нарисовать контурную карту функции $f(x, y)$ на выбранной самостоятельно прямоугольной области плоскости $x-y$.
3. добавить подписи к осям
4. добавить заголовок к рисунку
5. отобразить цветовую шкалу
6. cохранить изображение в формате `pdf` и разрешении `300 dpi`

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

func = X**3 + Y**3 + 2*(X - Y)* np.sin(X)


plt.figure(figsize = (10, 6))
contour_map = plt.contour(X, Y, func, levels = 50, cmap = 'plasma')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('Контурная карта к функции f(x, y) = x^3 + y^3 + 2(x - y)sin(x)', fontsize = 12)
plt.colorbar(contour_map, label='Значение функции f(x, y)')
plt.savefig('counter_map.pdf', dpi = 300)
plt.show()

````

## Задача Pandas 1 (10 баллов)

**Дано**

Таблица: [`stockholm_td_adj.dat`](https://github.com/jrjohansson/scientific-python-lectures/raw/master/stockholm_td_adj.dat).

1. Вывести на экран:
    - содержимое,
    - размеры,
    - типы элементов.

2. Дать имена столбцам:

```
Год   М  Д      Т1      Т2      Т3 D
1800  1  1    -6.1    -6.1    -6.1 1
1800  1  2   -15.4   -15.4   -15.4 1
1800  1  3   -15.0   -15.0   -15.0 1
1800  1  4   -19.3   -19.3   -19.3 1
1800  1  5   -16.8   -16.8   -16.8 1
```

3. Алгоритмически убедиться, что в столбце `D` все значения одинаковые. Выбросить лишний столбец `D`.

4. Вывести индексы строк, в которых есть отличия между столбцами `T1, T2 и T3`.

5. Создать столбец `T`, значения которого будут усредненными значениями `T1, T2 и T3`.

6. Рассчитать по столбцу `T` среднюю температуру:
    - для каждого года
    - для каждого месяца
    - для каждого месяца каждого года


````python
import pandas as pd
import numpy as np

#Все средние значения температур были округлены до 6 знаков после запятой!!!

url = 'https://raw.githubusercontent.com/jrjohansson/scientific-python-lectures/master/stockholm_td_adj.dat'
df = pd.read_csv(url, sep='\s+', header=None)

df.columns = ['Год', 'М', 'Д', 'T1', 'T2', 'T3', 'D']


print(df.head())
print("\nРазмер таблицы:", df.shape)
print("Типы элементов:\n", df.dtypes) 

if df['D'].nunique() == 1:
    print('\nВсе значения столбца D одинаковы и равны', df['D'].unique()[0])
    df = df.drop(columns = 'D')
else:
    print('\nВ стоблце D не все значения одинаковы')

differ = (df['T1'] != df['T2']) | (df['T1'] != df['T3']) | (df['T2'] != df['T3'])
print("\nИндексы строк с отличиями между столбцами T1, T2, T3:\n", df[differ].index.tolist()) 

df['T'] = df[['T1', 'T2', 'T3']].mean(axis = 1)
print("\nСтоблец усредненных температур из столбцов T1, T2, T3:\n", np.round(df['T'], 6))

avg_year_tmp = df.groupby('Год')['T'].mean()
print("\nСредние значения температуры по годам:\n", np.round(avg_year_tmp, 6))

avg_month_tmp = df.groupby('М')['T'].mean()
print("\nСредние значение температуры по месяцам:\n", np.round(avg_month_tmp, 6))

avg_tmp_of_each_month = df.groupby(['Год', 'М'])['T'].mean()
print("\nСреднее значение температуры по каждому месяцу каждого года\n", np.round(avg_tmp_of_each_month, 6)) 

    
````

# Задача Numpy* (15 баллов)

**Дано**

В файлах `nodes.npy` и `triangles.npy` содержатся массивы `Numpy`, которые можно загрузить при помощи функции `np.load`:
- `nodes` - массив (N, 2), содержащий координаты точек (float)
- `tris` - массив (M, 3), содержащий тройки индексов (i, j, k) такие, что точки с индексами i, j, k из массива `nodes` образуют треугольник.

**Требуется**

1. Вычислить минимальное и максимальное значение среди всех треугольников:
    1. длин сторон
    2. периметров
    3. площадей
    4. углов

**Оценивание**

- [5 баллов] допускаются любые циклы и их заменители
- [10 баллов] не допускаются никакие циклы и их заменители; использовать только возможности `Numpy`

````python
import numpy as np
import math

#Все значения были округлены до 6 знаков после запятой!!!

nodes = np.load('nodes.npy')      
tris = np.load('triangles.npy').astype(int)   

A = nodes[tris[:, 0]]  
B = nodes[tris[:, 1]]
C = nodes[tris[:, 2]]

a = np.linalg.norm(A - B, axis = 1)
b = np.linalg.norm(B - C, axis = 1)
c = np.linalg.norm(C - A, axis = 1)

all_sides = np.stack([a, b, c], axis = 1)

min_side = np.round(all_sides.min(), 6)
max_side = np.round(all_sides.max(), 6)

all_perimetrs = all_sides.sum(axis = 1)
min_perimetr = np.round(all_perimetrs.min(), 6)
max_perimetr = np.round(all_perimetrs.max(), 6)

half_perimetr = all_perimetrs / 2
area = np.sqrt(half_perimetr * (half_perimetr - a) *(half_perimetr - b)* (half_perimetr - c))
min_area = np.round(area.min(), 6)
max_area = np.round(area.max(), 6)

cosA = (c**2 + b**2 - a**2) / (2 * c * b) 
cosB = (a**2 + c**2 - b**2) / (2 * a * c)
cosC = (a**2 + b**2 - c**2)/ (2 * a * b)

all_angls = np.arccos(np.clip(np.stack([cosA, cosB, cosC], axis = 1), -1.0, 1.0))
min_angl = np.round(math.degrees(all_angls.min()), 6)
max_angl = np.round(math.degrees(all_angls.max()), 6)

print(f"Минимальная длина сторона = {min_side}")
print(f"Максимальная длина стороны = {max_side}")

print(f"\nМинимальный периметр = {min_perimetr}")
print(f"Максимальный периметр = {max_perimetr}")

print(f"\nМинимальная площадь = {min_area}")
print(f"Максимальная площадь = {max_area}")

print(f"\nМинимальный угол(в градусах) = {min_angl}")
print(f"Максимальный угол(в градусах) = {max_angl}")
````

7. выбросить из данных две самых больших траты и снова построить прогнозы

````python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression



df = pd.read_csv('fin_v2.csv', parse_dates = ["ВРЕМЯ"])

df = df.sort_values("ВРЕМЯ")
df["Дата"] = df["ВРЕМЯ"].dt.date
df["Год"] = df["ВРЕМЯ"].dt.year
df["Месяц"] = df["ВРЕМЯ"].dt.month


profits = df[df["СУММА"] >0]
avg_month_profit = profits.groupby(["Год", "Месяц"])["СУММА"].sum().groupby("Год").mean()

spendings = df[df["СУММА"] < 0]
avg_month_spendings = spendings.groupby(["Год", "Месяц"])["СУММА"].sum().groupby("Год").mean()

df["СУММА_НА_КАЖД_ДЕНЬ"] = df["СУММА"].cumsum()
everyday_check_state = df.groupby("Дата")["СУММА"].sum().cumsum()

plt.figure(figsize = (12, 6))
plt.plot(everyday_check_state.index, everyday_check_state.values, color = 'blue')
plt.title('Линия изменения счета по дням', fontsize = 12)
plt.xlabel('Даты')
plt.ylabel('Состояние счета')
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()

negative_balance_dates = everyday_check_state[everyday_check_state < 0].index

biggest_spendings = spendings.nsmallest(5, "СУММА")[["ВРЕМЯ", "КАТЕГОРИЯ", "СУММА"]]

everyday_check_state = everyday_check_state.reset_index()
everyday_check_state["Дата"] = pd.to_datetime(everyday_check_state["Дата"]) 
everyday_check_state["День"] = (everyday_check_state["Дата"] - everyday_check_state["Дата"].min()).dt.days
X = everyday_check_state[["День"]].values
Y = everyday_check_state["СУММА"].values

last_day = X.max()
upcoming_days = np.arange(last_day + 1, last_day + (365*2 + 1)).reshape(-1, 1)

linear_regression = LinearRegression().fit(X, Y)
linear_prediction = linear_regression.predict(upcoming_days)

kefs = np.polyfit(X.flatten(), Y, deg=2) 
poly_func = np.poly1d(kefs)

square_prediction = poly_func(upcoming_days.flatten())


plt.figure(figsize = (12, 6))
plt.plot(everyday_check_state["Дата"], Y, label ="Фактическое состояние счета", color = 'blue')
plt.plot(everyday_check_state["Дата"].min() + pd.to_timedelta(upcoming_days.flatten(), unit="D"), linear_prediction, label="Линейный прогноз", color = 'orange')
plt.plot(everyday_check_state["Дата"].min() + pd.to_timedelta(upcoming_days.flatten(), unit="D"), square_prediction, label='Квадратичный прогноз', color = 'green')
plt.title('Прогноз состояния счета на 2 года вперед', fontsize = 12)
plt.xlabel('Дата')
plt.ylabel('Баланс')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.tight_layout()
plt.show()


df_with_no_two_spendings = df.copy()
two_biggest_spendings = df_with_no_two_spendings[df_with_no_two_spendings["СУММА"] < 0].nsmallest(2, "СУММА")
df_with_no_two_spendings = df_with_no_two_spendings.drop(two_biggest_spendings.index)

state_without_two_days = df_with_no_two_spendings.groupby("Дата")["СУММА"].sum().cumsum().reset_index()
state_without_two_days["Дата"] = pd.to_datetime(state_without_two_days["Дата"]) 
state_without_two_days["День"] = (state_without_two_days["Дата"] - state_without_two_days["Дата"].min()).dt.days
X2 = state_without_two_days[["День"]].values
Y2 = state_without_two_days["СУММА"].values


linear_regression_2 = LinearRegression().fit(X2, Y2)

kefs_2 = np.polyfit(X2.flatten(), Y2, deg=2) 
poly_func_2 = np.poly1d(kefs_2)

upcoming_days_2 = np.arange(X2.max() + 1, X2.max() + (365*2+1)).reshape(-1, 1)

linear_prediction_2 = linear_regression_2.predict(upcoming_days_2)
square_prediction_2 = poly_func_2(upcoming_days_2.flatten())


plt.figure(figsize = (12, 6))
plt.plot(state_without_two_days["Дата"], Y2, label ="Фактическое состояние счета", color = 'blue')
plt.plot(state_without_two_days["Дата"].min() + pd.to_timedelta(upcoming_days_2.flatten(), unit="D"), linear_prediction_2, label="Линейный прогноз", color = 'orange')
plt.plot(state_without_two_days["Дата"].min() + pd.to_timedelta(upcoming_days_2.flatten(), unit="D"), square_prediction_2, label='Квадратичный прогноз', color = 'green')
plt.title('Прогноз состояния счета на 2 года вперед без двух самых больших трат', fontsize = 12)
plt.xlabel('Дата')
plt.ylabel('Баланс')
plt.legend(loc = 'upper left')
plt.grid(True)
plt.tight_layout()
plt.show()


````
