## Задача Numpy 1 (10 баллов)

Создать массивы и вывести на экран:
- содержимое
- форму
- размер занимаемой памяти

1. одномерный массив, содержащий действительные числа $\pi, \frac{2}{\pi}, \frac{\pi}{4}, \frac{8}{\pi}, \frac{\pi}{16}, ..., \frac{\pi}{1024}$
2. одномерный массив, содержащий константу `True`, если число делится на 3, но не делится на 6, иначе - `False` для чисел 1, ..., 101.
3. двумерный массив, содержащий аргументы комплексных чисел $x + iy$, где $x = 2k/3, y=5m/3, k=0,...,10, m=0,...,10$

Оценивание:
- [5 баллов] использовать генераторы списков
- [5 баллов] использовать только возможности `Numpy`

````python
import numpy as np


np.set_printoptions(suppress=True)


P = np.pi
arr1 = np.array([P, 2/P, P/4, 8/P, P/16, 32/P, P/64, 128/P, P/256, 512/P, P/1024], dtype=float)

print("1:\nМассив:", np.round(arr1, 4))
print("Форма:", arr1.shape)
print("Размер (в байтах):", arr1.nbytes)


arr2 = np.array([(x % 3 == 0 and x % 6 != 0) for x in range(1, 102)], dtype=bool)

print("\n2:\nМассив:", arr2)
print("Форма:", arr2.shape)
print("Размер (в байтах):", arr2.nbytes)


k_vals = [2 * k / 3 for k in range(11)]
m_vals = [5 * m / 3 for m in range(11)]

complex_grid = np.array([[complex(x, y) for y in m_vals] for x in k_vals])
arg_array = np.angle(complex_grid)

print("\n3:\nМассив аргументов:\n", np.round(arg_array, 4))
print("Форма:", arg_array.shape)
print("Размер (в байтах):", arg_array.nbytes)
````

1. Создать двумерный массив из $16$ псевдослучайных чисел:

- $x = sin(t)(e^{cos(t)} - 2cos(4t) + sin^5(\frac{t}{12}))$
- $y = cos(t)(e^{cos(t)} - 2cos(4t) + sin^5(\frac{t}{12}))$

Рассчитать $x(t)$ и $y(t)$ для не менее 5000 значений параметра $t$.
Интервал измнения параметра $t$ подобрать самостоятельно.

Визуализировать кривую, выполнив код:

Оценивание:

````python
import numpy as np 
import matplotlib.pyplot as plt 

t = np.linspace(0, 20*np.pi, 5000)

x = np.cos(t) * (np.exp(np.sin(t/3)) + np.tan(t/4) - np.sin(2*t)*np.cos(t/5)) 
y = np.sin(t) * (np.exp(np.cos(t/2)) - np.tan(t/6) + np.cos(3*t)*np.sin(t/4))

plt.figure(figsize = (10, 6))
plt.grid(True)
plt.plot(x, y)

````

## Задача Matplotlib 1 (10 баллов)

Самостоятельно выбрать 5 нелинейных функций так, чтобы их минимальные и максимальные значения на отрезке $[0, 10]$ находились примерно в одном диапазоне.

На одном рисунке:
- изобразить линейные графики функций, каждый своим цветом/стилем
- каждая линия должна иметь подпись на легенде (отобразить легенду)
- подписать оси
- сохранить рисунок в формате `PNG`


````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)

f1 = np.sqrt(x + 4)
f2 = np.log(x + 2) + 1
f3 = 0.5 * np.sin(x) + 3
f4 = 2 + np.exp(-x / 3)
f5 = 1/20 * ((x - 5)**2) + 2

plt.figure(figsize= (10, 6))
plt.plot(x, f1, label='f1(x) = sqrt(x + 4)', color='blue')
plt.plot(x, f2, label='f2(x) = ln(x + 2) + 1', color='green')
plt.plot(x, f3, label='f3(x) = 1/2sin(x) + 3', color='red')
plt.plot(x, f4, label='f4(x) = 2 + exp(-x/3)', color='orange')
plt.plot(x, f5, label='f5(x) = 1/20(x - 5)^2 + 2', color='purple')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.savefig('graphics.png')
plt.show()
````

````python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 



df = pd.read_csv("SolarPrediction.csv")

float_colls = df.select_dtypes(include='float').columns[:4]
col1, col2, col3, col4 = float_colls

plt.figure(figsize= (10, 6))
sc = plt.scatter(df[col1], df[col2], c = df[col3], s = df[col4]*10, cmap="viridis", alpha = 0.7)
plt.xlabel(col1)
plt.ylabel(col2)
plt.colorbar(sc, label = col3)
plt.legend([f'Размер точек соответствует {col4}'], loc = 'upper right')
plt.grid(True)


fig, axes = plt.subplots(2, 2, figsize=(10, 8))

columns = [col1, col2, col3, col4]
for i, ax in enumerate(axes.flat):
    ax.hist(df[columns[i]], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(f'Гистограмма {columns[i]}')
    ax.grid(True)

fig.suptitle('Гистограммы выбранных стоблцов', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])

````
